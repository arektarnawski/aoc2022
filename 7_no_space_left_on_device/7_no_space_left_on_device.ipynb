```python
# Part one
import pandas as pd
import re
pd.set_option('display.max_rows', 100)
```

```python
with open ('input.txt', 'rb') as input:
    df = pd.read_csv(input, names=['commands'])
df.head(20)
```

```python
df['level'] = ''
df['folder'] = ''
df['size'] = ''

for index, row in df.iterrows():
    command = row['commands']

    if command == '$ cd /':
        row['level'] = 0
        row['folder'] = '/'
        row['size'] = 0
    elif command == '$ ls' or re.search('dir ', command):
        row['level'] = df['level'][index-1]
        row['folder'] = df['folder'][index-1]
        row['size'] = 0
    elif re.search('\d+\s[a-zA-Z.]+', command):
        row['level'] = df['level'][index-1]
        row['folder'] = df['folder'][index-1]
        row['size'] = int(re.findall('\d+', command)[0])
    elif re.search('^\$ cd\s+[a-zA-Z0-9_/]+', command):
        try: 
            row['level'] = int(df['level'][index-1])+1
            row['folder'] = re.findall('[a-zA-Z0-9_/]+', command)[1]
            row['size'] = 0
        except:
            pass
    elif command == '$ cd ..':
        try: 
            row['level']= int(df['level'][index-1])-1
            row['folder'] = df['folder'][index-1]-1
            row['size'] = 0
        except:
            pass
# df['size'] = df['size'].astype(int)
df.head(15)
```
