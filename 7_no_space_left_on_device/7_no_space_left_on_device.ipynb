```python
# Part one
import pandas as pd
import re
pd.set_option('display.max_rows', 100)
```

```python
with open ('input.txt', 'rb') as input:
    df = pd.read_csv(input, names=['commands'])                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 d_csv(input, names=['commands'])
df.head(20)
```

```python
df['level'] = ''
df['folder'] = ''
df['size'] = ''
df['current_path'] = ''

current_path = '/'

for index, row in df.iterrows():
    command = row['commands']

    if command == '$ cd /':
        row['level'] = 0
        row['folder'] = '/'
        row['size'] = 0
        row['current_path'] = current_path
    elif command == '$ ls' or re.search('dir ', command):
        row['level'] = df['level'][index-1]
        row['folder'] = df['folder'][index-1]
        row['size'] = 0
        row['current_path'] = df['current_path'][index-1]
    elif re.search('\d+\s[a-zA-Z.]+', command):
        row['level'] = df['level'][index-1]
        row['folder'] = df['folder'][index-1]
        row['size'] = int(re.findall('\d+', command)[0])
        row['current_path'] = df['current_path'][index-1]
    elif re.search('^\$ cd\s+[a-zA-Z0-9_/]+', command):
        try: 
            row['level'] = int(df['level'][index-1])+1
            row['folder'] = re.findall('[a-zA-Z0-9_/]+', command)[1]
            row['size'] = 0
            row['current_path'] = df['current_path'][index-1] + re.findall('[a-zA-Z0-9_/]+', command)[1] + "/"
        except:
            pass
    elif command == '$ cd ..':
        try: 
            row['level']= int(df['level'][index-1])-1
            row['size'] = 0

            no_of_chars_to_cut = len(re.findall('\/[a-z]+/{1}$', df['current_path'][index-1])[0]) - 1
            trim_path = df['current_path'][index-1][:len(df['current_path'][index-1]) - no_of_chars_to_cut]
            row['current_path'] = trim_path
        except:
            pass

df['size'] = df['size'].replace('','0')
df['size'] = pd.to_numeric(df['size'])
df['level'] = pd.to_numeric(df['level'])
    
df.head(50)
```

```python
df_paths = df.groupby("current_path", as_index=False)["size"].sum()
```

```python
df_folders_and_paths = df[['folder','current_path']]
df_folders_and_paths.drop_duplicates(inplace=True)
df_folders_and_paths = df_folders_and_paths[df_folders_and_paths['folder']!='']
```

```python
df_folders_and_paths
```

```python
final_df = pd.DataFrame(columns=['folder', 'total_size'])
final_df

for index, row in df_folders_and_paths.iterrows():
    total_size = df_paths['size'][df_paths['current_path'].str.contains(row['current_path'])].sum()
    final_df = final_df.append({'folder': row['folder'], 'total_size': total_size}, ignore_index=True)
```
