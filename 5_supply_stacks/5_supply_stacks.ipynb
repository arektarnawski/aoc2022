```python
# Part one
import pandas as pd
import re
pd.set_option('display.max_rows', 100)
```

```python
# Reading in-the file for overview of the structure; removing the next-line character with .strip()
with open ('input.txt', 'rb') as f:
    for line in f:
        print(line.strip())
```

```python
# Reading in the first 8 lines containing the crates and appending to a list
counter = 0 
lines = []

with open('input.txt') as f:
    for line in f:
        if counter != 8:
            lines.append(line.rstrip())
            counter += 1
        else:
            break
lines
```

```python
# Assigning the numbers to the crates (1, 2, 3 ... up to 72). A blank space in the grid is also considered a crate at this point.
crates = {}
crate_no = 1
for line in lines:
    for i in range(9):
        crates[crate_no] = line[0+i*4:3+i*4]
        crate_no += 1
crates
```

```python
# Removing the square brackets from the crate letters
for key, value in crates.items():
    crates[key] = value.replace("[","").replace("]","")
crates
```

```python
# Creating a list of lists (list_of_rows) to mimic the columns of the original dataset - there are 9 lists ( = columns)
list_of_rows = []
for i in range(1,10):
    column = []
    list_of_rows.append(column)

# Appending the individual crates into their specific rows - so essentially this is a transposition
for key, value in crates.items():
    modulo = key % 9 - 1
    list_of_rows[modulo].append(value)

# Reversing the individual lists ( = columns) so that they mimic the stacked crates
for list in list_of_rows:
    list = list.reverse()

# Inserting an additional, empty list at index 0, so that we can later index starting from 1 and not 0
list_of_rows.insert(0,[])

# Removing the 'empty' crates to only have the real crates stacked in the lists ( = columns)
for list in list_of_rows:
    while "   " in list:
        list.remove("   ")
        
list_of_rows
```

```python
with open ('input.txt', 'rb') as input:
    df = pd.read_csv(input, skiprows=9, names=['moves'])
df
```

```python
# Formatting the df with all moves to extract the number of moves, the 'from' and the 'to' numbers
df['count'] = ''
df['from'] = ''
df['to'] = ''

for index, row in df.iterrows():
    row['count'] = re.findall('\d+', row['moves'])[0]
    row['from'] = re.findall('\d+', row['moves'])[1]
    row['to'] = re.findall('\d+', row['moves'])[2]
df.drop("moves", axis=1, inplace=True)
df
```

```python
# Final loop - using the list.pop() function to remove the 'count' number of elements from the 'from' list and append them to the 'to' list
for index, row in df.iterrows():
    _count = int(row['count'])
    _from = int(row['from'])
    _to = int(row['to'])
    
    for i in range(1,_count+1):
        list_of_rows[_to].append(list_of_rows[_from].pop())
list_of_rows

# The answer is simply the concat of characters in the at the end of each list - here it's 'RFFFWBPNS'
```

```python
# Part two
# All code is the same up until the final loop of part 1
```

```python
# Final loop - situation developed a little:
# (i) Still using the list.pop() function to remove the 'count' number of elements from the 'from' list 
# (ii) However they are now appended to a temp list first, which is then reversed - this mimics the behavior of lifting all crates at once instead of one by one
# (iii) The reversed temp_list is extended (not appended!) to the '_to' list (extending means addind one list's element to the other list; appending would add the entire list as an object to the other list)

for index, row in df.iterrows():
    _count = int(row['count'])
    _from = int(row['from'])
    _to = int(row['to'])

    temp_list = []
    for i in range(1,_count+1):
        temp_list.append(list_of_rows[_from].pop())
    temp_list.reverse()
    list_of_rows[_to].extend(temp_list)

list_of_rows

# The answer is simply the concat of characters in the at the end of each list - here it's 'CQQBBJFCS'
```
